#3. Which airline offers the lowest average ticket price for Business class?

```python
df[df['class']=='Business'].groupby('airline')['price'].mean().round(2).sort_values(ascending=True).head(1)
```

#4. Find routes with more than 50 flights and their average ticket prices.

```python
route_counts = df.groupby(['source_city', 'destination_city']).size().reset_index(name='No_of_flights')
routes_with_many_flights = route_counts[route_counts['No_of_flights'] > 50]
average_prices = df.groupby(['source_city', 'destination_city'])['price'].mean().reset_index(name='Average_Price')
result = pd.merge(routes_with_many_flights, average_prices, on=['source_city', 'destination_city'])
display(result)
```

# 5. Compare average price difference between Economy and Business class per airline

```python
average_price_per_class = df.groupby(['airline', 'class'])['price'].mean().unstack()
average_price_per_class['price_difference'] = average_price_per_class['Business'] - average_price_per_class['Economy']
display(average_price_per_class)
```

# 7. Which cities have the most outgoing flights?

# 8. Find the busiest route (most number of flights).

```python
most_busy_route=df.groupby(['source_city','destination_city']).size().reset_index(name='flights').sort_values('flights', ascending=False).head(1)
display(most_busy_route)
```

#9.List the top 3 airlines with cheapest average prices for each route.

```python
cheapest_airlines_per_route = df.groupby(['source_city', 'destination_city', 'airline'])['price'].mean().reset_index(name='average_price')
cheapest_airlines_per_route = cheapest_airlines_per_route.sort_values(['source_city', 'destination_city', 'average_price'])
top_3_cheapest = cheapest_airlines_per_route.groupby(['source_city', 'destination_city']).head(3)
display(top_3_cheapest)
```

#10.Find flights with duration more than 5 hours but priced below average

```python
average_price = df['price'].mean()
flights_long_and_cheap = df[(df['duration'] > 5) & (df['price'] < average_price)]
display(flights_long_and_cheap)
```

#11.11. Which flight class shows the highest average price variation (std deviation)?

# 12. Find how ticket prices vary with number of days left before departure.

# Visualizations

1. price distribution across air lines

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='airline', y='price')
plt.title('Price Distribution Across Airlines')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

2.	Average Price: Economy vs. Business Class

```python
average_price_per_class = df.groupby('class')['price'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(data=average_price_per_class, x='class', y='price',palette='Set2')
plt.title('Average Price: Economy vs. Business Class')
plt.xlabel('Class')
plt.ylabel('Average Price')
plt.show()
```

3.Busiest Source & Destination Cities

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=df, y='source_city', order=df['source_city'].value_counts().index, palette='viridis')
plt.title('Busiest Source Cities')
plt.xlabel('Number of Flights')
plt.ylabel('Source City')

plt.subplot(1, 2, 2)
sns.countplot(data=df, y='destination_city', order=df['destination_city'].value_counts().index, palette='viridis')
plt.title('Busiest Destination Cities')
plt.xlabel('Number of Flights')
plt.ylabel('Destination City')

plt.tight_layout()
plt.show()
```

4.Days Left vs. Ticket Price Trend

```python
price_variation_by_days_left = df.groupby('days_left')['price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=price_variation_by_days_left, x='days_left', y='price')
plt.title('Ticket Price Trend vs. Days Left Before Departure')
plt.xlabel('Days Left Before Departure')
plt.ylabel('Average Ticket Price')
plt.grid(True)
plt.show()
```

5.Stops vs. Price Comparison

```python
average_price_by_stops = df.groupby('stops')['price'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=average_price_by_stops, x='stops', y='price', order=['zero', 'one', 'two_or_more'])
plt.title('Average Ticket Price vs. Number of Stops')
plt.xlabel('Number of Stops')
plt.ylabel('Average Ticket Price')
plt.show()
```

6.Duration Impact on Ticket Pricing

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='duration', y='price', alpha=0.5)
plt.title('Ticket Price vs. Flight Duration')
plt.xlabel('Duration (hours)')
plt.ylabel('Ticket Price')
plt.grid(True)
plt.show()
```

```python
# Create duration bins
bins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
labels = [f'{i}-{i+5} hrs' for i in range(0, 50, 5)]
df['duration_binned'] = pd.cut(df['duration'], bins=bins, labels=labels, right=False)

# Create the boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='duration_binned', y='price')
plt.title('Ticket Price Distribution by Flight Duration Bins')
plt.xlabel('Duration Bins')
plt.ylabel('Ticket Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Here are 5 key findings from the analysis:

1.  **Vistara and Air India generally have higher average ticket prices** compared to other airlines, especially in Business class. SpiceJet and AirAsia tend to offer lower average prices.
2.  **Booking flights further in advance (more days left) generally leads to lower average ticket prices.** Prices tend to be higher when booking closer to the departure date.
3.  **Business class shows a significantly higher average price and price variation (standard deviation)** compared to Economy class.
4.  **The busiest routes in terms of the number of flights** are primarily between major cities like Delhi and Mumbai.
5. **Flights with more stops tend to have higher average prices** compared to non-stop flights, although this can vary by airline.
